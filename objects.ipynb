## Code for generating merged tickets & assets csv

```python
import pandas as pd
import numpy as np
from data_generator import df_convert_to_datetimes, merge_tickets_assets

df_tickets = pd.read_csv("data/V_OM_WORK_TASK.csv")
df_assets = pd.read_csv("data/V_OM_WORK_TASK_ASSET.csv")

# Generate merged tickets / assets file
df_tickets_assets = merge_tickets_assets(df_tickets=df_tickets, df_assets=df_assets)

df_tickets_assets["BASELINE_START_LTZ"].head()
```

```python
df_tickets_assets = df_tickets_assets.dropna(subset=['SPACE'])
# df_tickets_assets.head()

print(f"Number of rows with room number attached: {len(df_tickets_assets)}")

df_tickets_assets["SPACE"].head()
```

```python
# Clean tickets/assets data
df_tickets_assets['WORK_TASK_ID'] = df_tickets_assets['WORK_TASK_ID'].astype('Int64')

# Convert PLANNED_WORKING_DAYS to integer type
df_tickets_assets['PLANNED_WORKING_DAYS'] = df_tickets_assets['PLANNED_WORKING_DAYS'].astype('Int64')

# List of date columns to convert and format
date_columns = [
    'BASELINE_START_LTZ', 'BASELINE_END_LTZ', 'ASSIGNED_DATE_LTZ',
    'PLANNED_START_LTZ', 'PLANNED_END_LTZ', 'PLANNED_FOLLOW_UP_DATE_LTZ',
    'ACTUAL_START_LTZ', 'ACTUAL_END_LTZ', 'CREATE_DATE_LTZ'
]

# Convert and format date columns
for col in date_columns:
    # Ensure column is first converted to datetime, coercing errors
    df_tickets_assets[col] = pd.to_datetime(df_tickets_assets[col], errors='coerce')
    # Format datetime objects to the desired string format, leave NaT as NaN (which will be empty string in CSV)
    df_tickets_assets[col] = df_tickets_assets[col].dt.strftime('%Y-%m-%d %H:%M:%S.%f').replace({np.nan: ''})
```

## Rules for individual "object"

**All of the following conditions must hold:**
1. Same service class
2. Same request class
3. Same building 

**Future idea:**
1. Same keywords in description (cosine similarity score?)

If the same "object," as defined above, has multiple work tasks in the same 90-day window, then we flag it as a possible persistent issue (repetitive task).

```python
print(df_tickets_assets['BASELINE_START_LTZ'].dtype)

print(df_tickets_assets['BASELINE_START_LTZ'])
```

```python
# Convert to datetime
df_tickets_assets = pd.read_csv("data/df_merged_tickets_assets.csv")

df_tickets_assets['BASELINE_START_LTZ'] = pd.to_datetime(
    df_tickets_assets['BASELINE_START_LTZ'],
    errors='coerce'
)
df_tickets_assets.head()

# Sort the DataFrame by ASSET_ID and BASELINE_START_LTZ for efficient processing
df_merged_assets_sorted = df_tickets_assets.sort_values(by=['SERVICE_CLASS', 'REQUEST_CLASS', 'BUILDING', 'BASELINE_START_LTZ']).reset_index(drop=True)

repeated_tasks_by_object_list = []
ninety_days = pd.Timedelta(days=90)
min_days = pd.Timedelta(days=3) # Set minimum days between each repetitive task to avoid accidental double-submissions by users.

print(ninety_days)

# Filter by corrective first
df_corrective = df_merged_assets_sorted[
    df_merged_assets_sorted['TASK_TYPE'] == 'Corrective' 
]

df_corrective = df_merged_assets_sorted[
    df_merged_assets_sorted['BUILDING'] == 'Duncan College'
]

# FOR TESTING: Remove empty descriptions
df_corrective_2 = df_corrective[
    (df_corrective['DESCRIPTION'].notna()) &
    (df_corrective['DESCRIPTION'].str.strip() != '')
]

print(f"Number of corrective tasks: {len(df_corrective)}")

print_groups = False

# Group by "Object"
group_cols_1 = ['SERVICE_CLASS', 'BUILDING', 'FLOOR', 'SPACE'] # Try tasks with same description

unique_objects = df_corrective_2.groupby(group_cols_1).ngroups
print(f"Number of unique objects (groups): {unique_objects}")

for group_key, group in df_corrective_2.groupby(group_cols_1):
    if len(group) > 1 and print_groups:
        print("new group")
        print(group['DESCRIPTION'].tolist())
    # group_key is a tuple: (SERVICE_CLASS, REQUEST_CLASS, BUILDING)

    for i in range(len(group)):
        current_task = group.iloc[i]
        current_task_time = current_task['BASELINE_START_LTZ']

        # Check subsequent tasks for repetition within 90 days
        subsequent_tasks = group.iloc[i+1:]
        for j in range(len(subsequent_tasks)):
            other_task = subsequent_tasks.iloc[j]
            other_task_time = other_task['BASELINE_START_LTZ']

            time_difference = other_task_time - current_task_time

            if pd.isna(time_difference):  # Skip if time_difference is NaT
                continue

            if time_difference <= ninety_days and time_difference >= min_days:
                repeated_tasks_by_object_list.append(current_task.to_dict())
                break
            elif time_difference > ninety_days:
                break



# Create a new DataFrame from the identified repeated tasks
df_tickets_filtered_by_object = pd.DataFrame(repeated_tasks_by_object_list)

print(f"Found {len(df_tickets_filtered_by_object)} repetitive tasks by object within {ninety_days.days} days.")



```

```python
import matplotlib.pyplot as plt

group_sizes = df_corrective_2.groupby(group_cols_1).size()

size_distribution = group_sizes.value_counts().sort_index()
size_distribution.index.name = 'Group Size'
size_distribution.name = 'Number of Objects'

print(size_distribution.to_string())
```
