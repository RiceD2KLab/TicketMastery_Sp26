```python
import pandas as pd
import pydeck as pdk
import numpy as np
```

```python
df_1 = pd.read_csv(r'C:\Users\EC712\OneDrive\Documents\school\DSCI_435\data_sheets\TICKETS_WITH_COORDS.csv')
df_1 = df_1.dropna(subset=['FEP_BUILDING_X_COORDINATE', 'FEP_BUILDING_Y_COORDINATE'])
```

```python
df_1 = df_1[['FEP_BUILDING_X_COORDINATE', 'FEP_BUILDING_Y_COORDINATE', 'FEP_BUILDING_DESC', 'FEP_BUILDING_CLASS' ]]
df_1 = df_1.groupby(['FEP_BUILDING_Y_COORDINATE', 'FEP_BUILDING_X_COORDINATE',  'FEP_BUILDING_DESC', 'FEP_BUILDING_CLASS']).size().reset_index(name='COUNT')

colors = {0: [0, 0, 255], 1: [0, 255, 0], 2: [255, 255, 0], 3: [255, 127, 0], 4: [255, 0, 0]}

df_1['COLOR'] = pd.qcut(df_1['COUNT'], q=5, labels=False)
df_1['COLOR'] = df_1['COLOR'].map(colors)
```

```python
records = df_1.to_dict('records')
```

```python
# layer = pdk.Layer(
#     'HexagonLayer',
#     data=records,
#     get_position=['FEP_BUILDING_Y_COORDINATE', 'FEP_BUILDING_X_COORDINATE'],
#     auto_highlight=True,
#     elevation_scale=30,
#     pickable=True,
#     elevation_range=[0, 10],
#     extruded=True,
#     coverage=10,
#     radius=1)

layer = pdk.Layer(
  'ColumnLayer',
  data=records,
  diskResolution= 12,
  extruded=True,
  radius=10,
  elevationScale=0.025,
  get_position=['FEP_BUILDING_Y_COORDINATE', 'FEP_BUILDING_X_COORDINATE'],
  get_fill_color='COLOR',
  getElevation = 'COUNT',
  pickable=True
)

# Set the viewport location
view_state = pdk.ViewState(
    longitude=-95.404182,
    latitude=29.717154,
    zoom=15.5,
    min_zoom=15,
    max_zoom=20,
    pitch=45)

# Combined all of it and render a viewport
r = pdk.Deck(layers=[layer], initial_view_state=view_state, tooltip={"text": "Building: {FEP_BUILDING_DESC}\n Type: {FEP_BUILDING_CLASS}\n Tickets: {COUNT} "})
r.to_html("my_pydeck_map.html", open_browser=True)

```
